```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import mean_squared_error
```

```python
input_df=df2num.drop('pm10',axis=1)
output_df=df2num['pm10']
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
#20% dataset
x_train, x_test, y_train, y_test = train_test_split(input_df, output_df, test_size = 0.2, random_state = 0,shuffle=False)
LinReg_model  = LinearRegression()
LinReg_model .fit(x_train, y_train)
y_predict=LinReg_model.predict(x_test)

mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse)
print("Mean Squared Error:", mse)
print("Mean Squared Error:", rmse)
```

```python
#30% dataset
x_train, x_test, y_train, y_test = train_test_split(input_df, output_df, test_size = 0.3, random_state = 0,shuffle=False)
LinReg_model  = LinearRegression()
LinReg_model .fit(x_train, y_train)
y_predict=LinReg_model.predict(x_test)

mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse)
print("Mean Squared Error:", mse)
print("Mean Squared Error:", rmse)
```

Nilai error akan jauh lebih rendah jika testing set dibagi sebanyak 30%. Hal ini sudah kami coba untuk tipe prediction method lainnya, dan terbukti valid bahwa error dapat berkurang drastis untuk setiap test_size=0.3. Maka mulai dari sini, kami hanya akan menampilkan code dengan test_size=0.3 saja

```python
#adjusting parameter
LinReg_model  = LinearRegression(fit_intercept= False)
LinReg_model .fit(x_train, y_train)
y_predict=LinReg_model.predict(x_test)

mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse)
print("Mean Squared Error:", mse)
print("Mean Squared Error:", rmse)
```

dengan mencustom parameter, nilai error juga dapat ditekan

```python
#error of the training set
def find_error(df):
    error = {}

    for col in df.columns:
        df_copy = df.copy()
        input_df = df_copy.drop(col, axis=1)
        output_df = df_copy[col]
        x_train, x_test, y_train, y_test = train_test_split(input_df, output_df, test_size=0.3, random_state=0,shuffle=False)
        
        LinReg_model = LinearRegression(fit_intercept=False)
        LinReg_model.fit(x_train, y_train)
        y_predict = LinReg_model.predict(x_train)

        mse = mean_squared_error(y_train, y_predict)
        rmse = np.sqrt(mse)
       
        error[col] = {
            "Mean Squared Error": mse,
            "Root Mean Squared Error": rmse,
            "actual": y_train,
            "predicted": y_predict
        }

    return error

error_report = find_error(df2num)

for col, stats in error_report.items():
    print("Column:", col)
    print("Mean Squared Error:", stats['Mean Squared Error'])
    print("Root Mean Squared Error:", stats['Root Mean Squared Error'])
    print()
```
